```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('output.csv')
data
```

```python
nn = data['nn']
movielens_ubcf = data['mubcf_mea']/100
movielens_ubcfi = data['mubcfi_mea']/100

plt.figure(figsize=(10,5))

plt.plot(nn, movielens_ubcf, marker="x")
plt.plot(nn, movielens_ubcfi, marker="*")

plt.title('Mean Absolute Error - Accuracy with Movielens Dataset')
plt.xlabel('Size of Nearest Neighbor')
plt.ylabel('Accuracy')
plt.legend(['Movie Lens UBCF', 'Movie Lens UBCF-i'], loc='lower right')

plt.show()
```

```python
av_mubcf = data['mubcf_mea'].mean()
av_mubcfi = data['mubcfi_mea'].mean()

bar_data = {'Movielens-UBCF': av_mubcf, 'Movielens-UBCF-i': av_mubcfi}
head = list(bar_data.keys())
value = list(bar_data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(head, value, width=0.8, align='center')

plt.xlabel("Algorithms Tested")
plt.ylabel("Average Accuracy")
plt.title("Overall Average MAE Accuracy of UCBF & UBCF-i with Movie Lens Dataset")
plt.show()
```

```python
nn = data['nn']

jester_ubcf = data['jubcf_mea']
jester_ubcfi = data['jubcfi_mea']

plt.figure(figsize=(10,5))

plt.plot(nn, jester_ubcf, marker="o")
plt.plot(nn, jester_ubcfi, marker="d")

plt.title('Mean Absolute Error - Accuracy with Jester Dataset')
plt.xlabel('Size of Nearest Neighbor')
plt.ylabel('Accuracy')
plt.legend(['Jester UBCF', 'Jester UBCF-i'], loc='lower right')

plt.show()
```
